```python
def deduplication(x):
    mid = len(x)//2
    if x!='unknown':
        return x[:mid + 1]
    else:
        return 'unknown'
```

```python
def deduplication_better(x):
    trim = x.strip()
    index = (trim + trim).find(trim,1)
    if index != -1:
        return trim[:index]
    else:
        return 'unknown'
```

```python
df_admin['location'] = df_admin['location'].apply(deduplication_better)
```
